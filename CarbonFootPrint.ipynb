##Problems with the data quality:

*   Some rows are aggregates of countries rather than actual countries (e.g., "World").
*   Some columns are irrelevant and can be removed (e.g., "Indicator Name").
*   Some years have no data for any country (e.g., 2012 to 2015).
*   Some countries have no data for any year. (e.g., "Taiwan, China")

##Solution :
**Some rows are aggregates of countries rather than actual countries (e.g., "World").**
##Goals:
**Remove rows that do not contain an actual country. Fortunately, the World Bank provides us with metadata on which rows are countries and which are aggregates.**

1.   import countries_metadata.csv
2.   merge metadata with data on Country Code

**Import countries_metadata.csv**

```python
metadata =pd.read_csv("/content/sample_data/country_metadata.csv",encoding = "utf-8")
```

```python
metadata.head(10)
```

**Identifing listed "Country Name" is a country or an aggregated region**

Notice when the row is an aggregate like "Arab World", the Region and IncomeGroup are consistently NaN (Not a Number). We can use this rule to remove all non-country regions.

**Merge data with metadata on the key, Country Code**

```python
merge = pd.merge(data,metadata, on = "Country Code")
merge.shape
```

```python
merge.head(10)
```

Removing the rows where Region is NaN, means the row is not a actual country

```python
merge = merge[pd.notnull(merge['Region'])]
```

```python
merge.head(2)
```

```python
merge.columns
```

**Some columns are irrelevant and can be removed**

Remove the following irrelevant columns:

Column 3: "Indicator Name"
Column 4: "Indicator Code"

*   List item
*   List item



```python
merge = merge.drop('Indicator Name', 1)
merge = merge.drop('Indicator Code',1)
```

```python
merge.columns
```

```python
merge.count()
```

Checking year 2020

```python
merge['2020']
```

```python
merge.mean(axis=1)
```

##Some countries have no data for any year.
Use row means to determine which countries have no data.

Let's take the means for each row (on axis 1).

**Now we can see that, the NaN appears when there is no data for that row.**

R**emove rows where there are no values in any year**

```python
merge=merge[pd.notnull(merge.mean(axis=1))]
```

```python
merge.head(45)
```

**Testing the search function, filter_country**

```python
def filter_country(search_term):
    '''
    Filters the medals_df DataFrame to only contain rows whose Sport or Discipline
    columns contain the given search_term.
    '''
    return mydf[mydf["Country Name_y"].str.contains(search_term, case=False) \
                     | mydf["Country Code"].str.contains(search_term, case=False)\
                     | mydf["Region"].str.contains(search_term, case=False)]
```
